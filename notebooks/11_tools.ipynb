```python
import pandas as pd
from pandas.plotting import scatter_matrix
```

```python
def get_numerical_features(df):
    numerical_features = []

    for feature, dtype in zip(df.columns, df.dtypes):
        if dtype == "int64" or dtype == "float64":
            numerical_features.append(feature)

    return numerical_features
```

### Get results

```python
params = pd.read_csv("/home/brandon/tfcaidm-pkg/examples/projects/pneumonia/exp/xr_pna/csvs/2021-11-13_23-36-28_PST_hyper.csv")
results = pd.read_csv("/home/brandon/tfcaidm-pkg/examples/projects/pneumonia/exp/xr_pna/csvs/2021-11-13_23-36-28_PST_results.csv")

params = params.dropna(axis=1)
results = results.dropna(axis=1)
```

```python
numerical_features = get_numerical_features(params)
target = ["valid_eval/fold/avg/last/dice_1"]
```

```python
subset = results[numerical_features + target]
```

### Correlation

Linear correlation assuming hyperparameters are independent of each other when considering performance.

In general, this is not true!

```python
corr = subset.corr()
pd.DataFrame(corr["valid_eval/fold/avg/last/dice_1"].sort_values(ascending=False))
```

**Note** a value of `NaN` means that the given parameters did not vary.

```python
_ = scatter_matrix(results[["model/elayer", "model/depth", "model/width"]])
```

### Random forest feature importance

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train = results[numerical_features].to_numpy()
y_train = results[target].to_numpy().reshape(-1)

forest = RandomForestRegressor(random_state=0)
forest.fit(X_train, y_train)
```

```python
forest.feature_importances_
```

```python
pd.DataFrame({k: v for k, v in zip(numerical_features, forest.feature_importances_)}.items())
```
